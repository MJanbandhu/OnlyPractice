# Hypothesis Testing

# ZTest

```python
# There is a claim that average age of IT male professionals in Bangalore is 35 years.
# H0:average age is 35 years
# Ha:average age is not 35 years
```

```python
# z test
data=[36,25,26,27,33,32,37,27,46,28,29,25,27,28,34,35,36,37,28,24,29,32,31,26,27,28]
```

```python
from statsmodels.stats.weightstats import ztest
```

```python
p,z= ztest(data,value=35)
```

```python
print("P is",p,"Z is",z)
```

```python
# At 5% level of significance, we reject null hypothesis because p<alpha.
# We accept alternate hypothesis
```

```python
print(f"P is {p} And Z is {z}")
```

```python
print("P is{} and Z is {}".format(p,z))
```
