# Python program to find mean of given data.

```python
data = [10, 20, 30, 40, 50]
mean = sum(data) / len(data)
print("Mean =", mean)
```

# Python program to generate Fibonacci series up to 100.

```python
a, b = 0, 1
while a <= 100:
    print(a, end=" ")
    a, b = b, a + b
```

# Python program to find factorial of a number.

```python
num = 5
fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial =", fact)
```

# Program to Check Prime Number

```python
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a Prime Number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")
```
