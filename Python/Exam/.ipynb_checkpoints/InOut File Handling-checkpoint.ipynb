# I/O File handling or Operation

I/O file operations in Python are handled through the built-in
open() function, which creates a file object used to interact with external files. This object provides methods for reading data from files, writing data to them, and managing file access. 
The open() Function



The fundamental function for all file operations is open(). It takes two primary arguments: filename (path to the file) and mode (how the file should be opened)

### Opening a file using open()

### Reading data using read(), readline()

### Writing data using write()

### Closing file using close()

```python
data= open('data.txt','w')
data.write("Hello Python")
data.close()
```

```python
data
```

# Mode 	Meaning	Description
      'r'	Read	Opens a file for reading. (Default mode)
    'w'	Write	Opens a file for writing. Creates a new file if it does not exist or truncates (clears) the file if it exists.
    'a'	Append	Opens a file for appending data to the end of the file. Creates a new file if it does not exist.
    'x'	Exclusive Creation	Creates a new file and opens it for writing. Raises an error if the file already exists.
    't'	Text	Opens in text mode. (Default mode)
    'b'	Binary	Opens in binary mode (for non-text files like images or executables).
    '+'	Update	Opens a file for updating (reading and writing).

```python
data= open('data.txt','r')

contend= data.read()


data.close()
```

```python
contend
```

For more advanced file system operations (like renaming, deleting, or checking file existence), Python provides the os and shutil modules. 

    os.remove(filename): Deletes a file.
    os.rename(old_name, new_name): Renames a file.
    os.path.exists(path): Checks if a path exists.
