```python
moh="Mohit Janbandhu 1234"
```

```python
moh="Mohit Janbandhu 1234"
vowels="AEIOUaeiou"
upper_case=0
lower_case=0
digits=0

for char in moh:
    if char.isupper():
        upper_case +=1
        
    elif char.islower():
        lower_case +=1
        
    elif char.isdigit():
        digits += 1
        
print("This much has Upper Latter{}".format(upper_case))
print(f" This mus has Lower digits {lower_case}")
print(f" This mush has dgits {digits}")
```
