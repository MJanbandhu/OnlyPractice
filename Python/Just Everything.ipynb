```python
moh="Mohit Janbandhu 1234"
```

```python
upper_latter=0
lower_latter=0
digits=0


for char in moh:
    if char.isupper():
        upper_latter +=1
        
        
    elif char.islower():
        lower_latter +=1
        
    elif char.isdigit():
        digits+=1
```

```python
print("Upper Latters are", upper_latter)
print(f"Lower Latter are {lower_latter}")
print("Digits are {}".format(digits))
```

```python
vowels="AEIOUaeiou"
for char in moh:
    if char in vowels:
        print(char)
```

```python
upper_latter=0
lower_latter=0
digits=0

upper_al=[]
lower_al=[]
digits_al=[]

for char in moh:
    if char.isupper():
        upper_latter +=1
        upper_al.append(char)
        
        
    elif char.islower():
        lower_latter +=1
        lower_al.append(char)
        
    elif char.isdigit():
        digits+=1
        digits_al.append(char)
        
print("Upper Latters are", upper_latter)
print(f"Lower Latter are {lower_latter}")
print("Digits are {}".format(digits))


print("Upper Alphabets are", upper_al)
print(f"Lower Alphabets are {lower_al}")
print("Digits are {}".format(digits_al))
```
