# Input

```python
# User enters numbers separated by space
inp = list(map(int, input("Enter numbers separated by space: ").split()))
print(inp)
```

```python
inp = list(map(int, input("Enter numbers separated by space: ").split()))
print(inp)
```

```python
# Take input number by number using a loop
n = int(input("How many numbers? "))
inp = []

for i in range(n):
    num = int(input(f"Enter number {i+1}: "))
    inp.append(num)

print(inp)
```

```python
# If user enters each digit without space (like 12345)
inp = [int(x) for x in input("Enter digits: ")]
print(inp)
```

# Area of triangle

```python
def area_triangle(base, height):
    return 0.5 * base * height

# Example usage
print(area_triangle(5, 10))  # Output: 25.0
```
