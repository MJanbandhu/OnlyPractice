```python
upper_case= 0
lower_case=0
digits=0

upper_letter=[]
lower_letter=[]
digits_all=[]
```

```python
for char in mohit:
    if char.isupper():
        upper_case +=1
        upper_letter.append(char)
        
        
    elif char.islower():
        lower_case +=1
        lower_letter.append(char)
        
        
        
    elif char.isdigit():
        digits +=1
        digits_all.append(char)
```

```python
upper_letter
```

```python
print("Upper case letter count is", upper_case)
print("Lower case letter count is{}".format(lower_case))
print(f"digit count is {digits}")
```
