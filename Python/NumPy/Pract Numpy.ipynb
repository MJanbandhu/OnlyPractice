## Built-in Methods

There are lots of built-in ways to generate Arrays

```python
### Reshape function
```

```python
arr1=np.array([1,2,3])
```

```python
arr2=arr1.reshape((1,3,4))  # reshape function takes the new shape as its argument
```

```python
arr2
```

```python
arr2.shape
```

```python
arr3 = arr1.reshape((3, 1))  # valid
```

```python
arr3
```

```python
arr4 = arr3.reshape((3, 3,1))  # valid
arr4
```

```python
arr2=arr1.reshape((1,3,4))  # reshape function takes the new shape as its argument
```

```python
arr5=np.reshape(4,2,3)
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
arr5 = np.reshape(arr, (2, 3))
print(arr5)
```
