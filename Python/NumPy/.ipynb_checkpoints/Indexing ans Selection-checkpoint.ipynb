# NumPy Indexing and Selection

Here we will discuss how to select elements or groups of elements from an array.

```python
import numpy as np
```

```python
arr_2d = np.array(([1,2,3],[12,15,18],[64,96,128]))

#Show
arr_2d
```

```python
arr_2d[1:,:2]
```

```python
arr_2d[1:,:1]
```

```python
arr_2d[1:,1:]
```

```python
arr_2d[1:2,1:2]
```
