#### Short Upper latter, Lower Latter and digits

```python
moh
```

```python
upper_latter=0
lower_latter=0
digits=0

for char in moh:
    if char.isupper():
        upper_latter +=1
        
    elif char.islower():
        lower_latter +=1
        
    elif char.isdigit():
        digits +=1
```

```python
print(upper_latter)
print(lower_latter)
print(digits)
```
