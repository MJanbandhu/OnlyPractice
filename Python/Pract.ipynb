# Variables

```python
# integer
intvar = 1000
print(intvar)
print(type(intvar))
print(sys.getsizeof(intvar))
print(isinstance(intvar,int))
print(isinstance(intvar,float))
```

```python
# Float
floatvar = 10.23234242423423
print(floatvar)
print(type(floatvar))
print(sys.getsizeof(floatvar))
print(isinstance(floatvar,float))
```

```python
# Boolean

bool1 = True
bool2 = False
print(type(bool1))
print(isinstance(bool1,bool))
print(isinstance(bool1,int))
print(isinstance(bool2,bool))
print(isinstance(bool2,int))
```

```python
print(bool("apple"))
print(bool("banana"))
print(bool(34))
print(bool(45.35))
print(bool(0))
print(bool(""))
print(bool('adasadsafwsdfabjhsdbjasfb'))
```

```python
## string

strvar = "apple"
print(strvar)
print(type(strvar))
print(sys.getsizeof(strvar))
print(isinstance(strvar,str))
```

```python
## String indexing
myvar = "Hello World"
# fetch 'W' from myvar
print(myvar[6])
print(myvar[-5])
# fetch 'r' from myvar
print(myvar[8])
print(myvar[-3])
# fetch the last character of myvar
print(myvar[-1])
```

```python
print(myvar)
print(len(myvar))
print(myvar.count('l'))
print(myvar.startswith("He"))
print(myvar.startswith("Hell"))
print(myvar.startswith("err"))
print(myvar.endswith("ld"))
print(myvar.index("e"))
print(myvar.index('l'))
```
