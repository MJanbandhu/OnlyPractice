### User enters numbers separated by space

```python
# User enters numbers separated by space
inp = list(map(int, input("Enter numbers separated by space: ").split()))
print(inp)
```

```python
inp = list(map(int, input("Enter numbers separated by space: ").split()))
print(inp)
```

```python
# Take input number by number using a loop
n = int(input("How many numbers? "))
inp = []

for i in range(n):
    num = int(input(f"Enter number {i+1}: "))
    inp.append(num)

print(inp)
```

```python
# If user enters each digit without space (like 12345)
inp = [int(x) for x in input("Enter digits: ")]
print(inp)
```

# Filter Function


The filter() method filters the given sequence with the help of a function that tests each element in the sequence to be true or not.

# Map Function

map() function returns a map object(which is an iterator) of the results 
after applying the given function to each item of a given 
iterable (list, tuple etc.)


# Reduce Function


The reduce(fun,seq) function is used to apply a particular function passed in
its argument to all of the list elements mentioned in the sequence passed 
along.This function is defined in “functools” module.
