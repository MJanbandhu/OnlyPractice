```python
list1 = [1, 5, 8, 4, 6, 9, 7, 4, 7]
def ev_odd():
    for char in list1:
        if char%2==0:
            print("Even")
            
        else:
            print("Odd")
```

```python
def ev_odd(number):
    for char in number:
        if char % 2==0:
            print("even")
            
        else:
            print('odd')
```
