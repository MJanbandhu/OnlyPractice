# Polymorphism

```python

class Man:
    def walk(self):
        print("Man walks fast")

class Woman:
    def walk(self):
        print("Woman walks slowly")

def show_walk(person):
    person.walk()

show_walk(Man())
show_walk(Woman())
```

```python
# Abstraction + Encapsulation
class Human:
    def __init__(self, name):
        self.name = name        # Encapsulation: variable inside class
    
    def hair_type(self):
        pass  # Abstract behavior (no implementation here)

# Inheritance + Polymorphism
class Male(Human):
    def hair_type(self):
        print(self.name, "has Short Hair")  # Polymorphism (different implementation)

class Female(Human):
    def hair_type(self):
        print(self.name, "has Long Hair")   # Polymorphism (different implementation)

# Object Instantiation
person1 = Male("John")
person2 = Female("Sara")

# Method Calls
person1.hair_type()
person2.hair_type()
```

| OOP Feature   | Where it appears                                                  |
| ------------- | ----------------------------------------------------------------- |
| Encapsulation | Variables and methods inside class (`self.name`, `hair_type`)     |
| Abstraction   | `hair_type()` in parent class has no detail (hidden logic)        |
| Inheritance   | `Male(Human)` and `Female(Human)` extending Human class           |
| Polymorphism  | Same method `hair_type()` behaves differently in each child class |

# Inheritance and method overriding

```python
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

Dog().sound()
```

# Encapsulation using private variable

```python
class Student:
    def __init__(self):
        self.__marks = 90  # Private variable

    def get_marks(self):
        return self.__marks

s = Student()
print(s.get_marks())
```

# Polymorphism example

```python
class A:
    def show(self):
        print("Class A show")

class B:
    def show(self):
        print("Class B show")

def display(obj):
    obj.show()

display(A())
display(B())
```

# Abstraction using abc

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def __init__(self, s):
        self.s = s

    def area(self):
        return self.s * self.s

print(Square(4).area())
```
