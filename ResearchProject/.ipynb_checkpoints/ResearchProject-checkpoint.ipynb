# October 2025 World Economic Outlook.

The World Economic Outlook (WEO) database is published twice a year in April and October in conjunction with the biannual flagship World Economic Outlook report. The database includes selected data on national accounts, inflation, unemployment rates, balance of payments, fiscal indicators, trade for countries and country groups (aggregates), and commodity prices whose data are reported by the IMF. Data are available from 1980 to the present, and projections for most data series are provided for the next five years. For some countries, country groups, and indicators, data are incomplete or unavailable for certain years.
