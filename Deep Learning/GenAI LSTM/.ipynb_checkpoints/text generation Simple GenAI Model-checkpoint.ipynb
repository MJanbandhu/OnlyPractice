# character-level text generation model by using (LSTM)

# 1. Problem Statement

Objective:
Build a Generative AI model that learns patterns from a given text dataset and generates new text similar to it.

Example use case:
Given a dataset of sentences (stories / articles), generate new text in the same style.

# 2. Dataset (Any Text Dataset)

# 3. Import Required Libraries

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

# 4. Load Dataset

```python

# Load text file
with open("data.txt", "r", encoding="utf-8") as file:
    text = file.read().lower()
```

# 5. Tokenization

Convert words into integer sequences.

```python

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

total_words = len(tokenizer.word_index) + 1
```

# 6. Create Input Sequences

```python

input_sequences = []

for line in text.split("\n"):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_seq = token_list[:i+1]
        input_sequences.append(n_gram_seq)
```

# 7. Padding Sequences

```python

max_len = max(len(seq) for seq in input_sequences)

input_sequences = pad_sequences(
    input_sequences,
    maxlen=max_len,
    padding="pre"
)
```

# 8. Split Input and Output

```python

X = input_sequences[:, :-1]
y = input_sequences[:, -1]

y = np.eye(total_words)[y]   # One-hot encoding
```

# 9. Build Generative Model (LSTM)

```python

model = Sequential([
    Embedding(total_words, 50, input_length=max_len-1),
    LSTM(100),
    Dense(total_words, activation="softmax")
])

model.compile(
    loss="categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)
```

# 10. Train the Model

```python

model.fit(X, y, epochs=100, verbose=1)
```

# 11. Text Generation Function

```python

def generate_text(seed_text, next_words):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_len-1, padding="pre")
        predicted = model.predict(token_list, verbose=0)
        predicted_word = tokenizer.index_word[np.argmax(predicted)]
        seed_text += " " + predicted_word
    return seed_text
```

# 12. Generate New Text

```python

print(generate_text("machine learning", 5))
```

```python
print(generate_text("machine learning", 6))
```

1. Markov Chain (Very Simple GenAI)
What it does

Generates text by predicting the next word based only on the previous word(s).

Used for

Simple text generation

Educational GenAI demos

No deep learning needed

When to use

Small dataset

Low computation

Concept demonstration

2. RNN (Recurrent Neural Network)
What it does

Processes sequences by remembering previous information.

Used for

Text generation

Sequence modeling

When to use

Small sequences

Simple generative tasks

Limitation

Suffers from vanishing gradient problem

3. GRU (Gated Recurrent Unit)
What it does

Improved version of RNN with fewer gates than LSTM.

Used for

Text generation

Time-series generation

When to use

Faster training than LSTM

Medium-sized datasets

4. CNN for Text Generation
What it does

Uses convolution filters to capture local word patterns.

Used for

Sentence generation

Text classification + generation hybrids

When to use

Short text

Pattern-based text generation

5. Autoencoder (AE)
What it does

Learns compressed representations and reconstructs data.

Used for

Text reconstruction

Feature generation

When to use

Data compression

Noise removal

6. Variational Autoencoder (VAE)
What it does

Generates new data by sampling from a learned distribution.

Used for

Text generation

Image generation

Data augmentation

When to use

Need diversity in generated output

Controlled randomness

7. GAN (Generative Adversarial Network)
What it does

Two models compete:

Generator

Discriminator

Used for

Image generation

Synthetic data generation

When to use

Creating realistic synthetic data

Note

GANs are rarely used for text because text is discrete.

8. N-Gram Language Model (No Neural Network)
What it does

Predicts next word using n previous words.

Used for

Text generation

Language modeling basics

When to use

Exam answers

Very simple implementations

9. Transformer (Non-GPT Implementation)
What it does

Uses self-attention, not recurrence.

Used for

Advanced text generation

Summarization, translation

When to use

Large datasets

High-performance GenAI

| Task                     | Best Algorithm |
| ------------------------ | -------------- |
| Simple text generation   | Markov Chain   |
| Small dataset GenAI      | N-gram         |
| Sequential text          | RNN            |
| Efficient sequence model | GRU            |
| Long-term dependencies   | LSTM           |
| Controlled generation    | VAE            |
| Image generation         | GAN            |
| State-of-the-art NLP     | Transformer    |
